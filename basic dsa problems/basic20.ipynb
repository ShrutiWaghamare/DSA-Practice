```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
op = input("Enter operator (+, -, *, /): ")

if op == '+':
    print(a + b)
elif op == '-':
    print(a - b)
elif op == '*':
    print(a * b)
elif op == '/':
    print(a / b)
else:
    print("Invalid operator")
```

```python
num = int(input("Enter the enumber: "))
fact = 1
for i in range(1, num + 1):
    fact *= i
print("factorial: ", fact)
```
