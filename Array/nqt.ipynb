```python
n = int(input("Enter a number: "))
rev = 0
while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n = n // 10
print("Reversed number: ", rev)
```

```python
# Second Maximum Element in Array
arr = list(map(int, input().split()))
lar = -1
seclar = -1
for i in range(len(arr)):
    if arr[i] > lar:
        seclar = lar
        lar = arr[i]
    elif arr[i] > seclar and arr[i] < lar:
        seclar = arr[i]
print(seclar)
```

```python
# Maximum Sum Subarray – Kadane’s Algorithm
arr = list(map(int, input().split()))
cur_sum = max_sum = arr[0]
for i in range(1, len(arr)):
    cur_sum = max(arr[i], cur_sum + arr[i])
    max_sum = max(max_sum, cur_sum)
print(max_sum)
```

```python
# how many Sundays occur within N days starting from a given weekday.
day = input().lower()
n = int(input())
days = {"mon":1, "tue":2, "wed":3,
    "thu":4, "fri":5, "sat":6, "sun":7}
start = days[day]
days_to_sunday = 7 - start
if days_to_sunday >= n:
    print(0)
else:
    count = 1 + (n-days_to_sunday-1)//7
    print(count)
```

```python
arr = list(map(int, input().split()))
low = mid = 0
high = len(arr)-1
while mid <= high:
    if arr[mid] == 0:
        arr[low], arr[mid] = arr[mid], arr[low]
        mid += 1
        low += 1
    elif arr[mid] == 1:
        mid += 1
    else:
        arr[mid], arr[high] = arr[high], arr[mid]
        high -= 1
print(arr)
```

```python
# Longest Substring Without Repeating Characters
s = "abcabcbb"
seen = set()
left = 0
length = 0
for right in range(len(s)):
    while s[right] in seen:
        seen.remove(s[right])
        left += 1
    seen.add(s[right])
    length = max(length, right - left + 1)
print(length)
```

```python
# Maximum Number of 'a' in Curtain Boxes

s = input()
L = int(input())
max_a = 0
for i in range(0, len(s), L):
    group = s[i:i+L]
    count_a = group.count('a')
    max_a = max(max_a, count_a)
print(max_a)
```

Circular Seating Arrangement Problem
Problem

There are N members sitting around a circular table.

Condition:

President and Prime Minister must sit next to each other.

Find the total number of possible seating arrangements.

```python
n = int(input())
fact = 1
for i in range(1, n):
    fact *= i
print(fact * 2)
```
