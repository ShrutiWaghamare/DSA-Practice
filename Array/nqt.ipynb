```python
# Count Positive, Negative, and Zero Numbers in Array
n = int(input())
arr = list(map(int, input().split()))
pos = neg = zero = 0
for num in arr:
    if num > 0:
        pos += 1
    elif num < 0:
        neg += 1
    else:
        zero += 1
print(pos, neg, zero)
```

```python
# Reverse Digits of a Number
n = input()
print(int(n[::-1]))
```

```python
n = int(input("Enter a number: "))
rev = 0
while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n = n // 10
print("Reversed number: ", rev)
```

```python
# Second Maximum Element in Array
arr = list(map(int, input().split()))
lar = -1
seclar = -1
for i in range(len(arr)):
    if arr[i] > lar:
        seclar = lar
        lar = arr[i]
    elif arr[i] > seclar and arr[i] < lar:
        seclar = arr[i]
print(seclar)
```

```python
# Maximum Sum Subarray – Kadane’s Algorithm
arr = list(map(int, input().split()))
cur_sum = max_sum = arr[0]
for i in range(1, len(arr)):
    cur_sum = max(arr[i], cur_sum + arr[i])
    max_sum = max(max_sum, cur_sum)
print(max_sum)
```

```python
# Star Pattern (Right Triangle)
n = int(input())
for i in range(1, n+1):
    print("*" * i)
```

```python
# move zeros to the end
arr = list(map(int, input().split()))
pos = 0
for i in range(len(arr)):
    if arr[i] != 0:
        arr[i], arr[pos] = arr[pos], arr[i]
        pos += 1
print(arr)
```
