Array:

[64, 25, 12, 22, 11]
Step-by-Step
Pass 1

Find smallest in whole array → 11

Swap with first element:

[11, 25, 12, 22, 64]
Pass 2

Find smallest from remaining → 12

[11, 12, 25, 22, 64]
Pass 3

Find smallest → 22

[11, 12, 22, 25, 64]
Pass 4

Find smallest → 25

[11, 12, 22, 25, 64]
Final Sorted Array
[11, 12, 22, 25, 64]

```python
# selection sort
arr = [64, 25, 12, 22, 11]
n = len(arr)

for i in range(len(arr)):
    min_ind = i
    for j in range(i+1, len(arr)):
        if arr[j] < arr[min_ind]:
            min_ind = j 
    arr[i], arr[min_ind] = arr[min_ind], arr[i]
print(arr)
```

```python
arr = [64, 25, 12, 22, 11]
for i in range(len(arr)):
    for j in range(i+1, len(arr)):
        if arr[i] > arr[j]:
            arr[i], arr[j] = arr[j], arr[i]
print(arr)
```

```python
arr = [64, 25, 12, 22, 11]
for i in range(1, len(arr)):
    key = arr[i]
    j = i - 1
    while j >= 0 and arr[j] > key:
        arr[j+1] = arr[j]
        j-= 1
    arr[j+1] = key
print(arr)
```
